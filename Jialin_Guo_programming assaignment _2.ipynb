```python
import matplotlib.pyplot as plt
```

```python
skills = ['General Programming', 'Python', 'Visualization', 'Graphic Design', 'Data Anlytics']
my_rate_points = [6, 6, 5, 7, 5.5]
```

```python
plt.xticks(rotation=90)
plt.bar(skills, my_rate_points)
plt.show()
```

The data about rating my experience are sort of seperate among these skills.These data don't have time-varying properties. Rather, they are seperate in each own category shows my rating at this point. So I think bar chart is a good choice to depict this relationship among them.
